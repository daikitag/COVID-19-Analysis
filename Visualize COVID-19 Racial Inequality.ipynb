# Part 2: Visualize COVID-19 Racial Inequality

```python
import numpy as np
import matplotlib.pyplot as plt
```

This data is taken from:

- National Center for Health Statistics

https://www.cdc.gov/nchs/nvss/vsrr/covid19/health_disparities.htm Accessed Nov 3 2020


```python
# These are COVID-19 death counts broken down by race
covid = np.array([111469, 43284,2179,8653,44359,2384])
race = np.array(['White','Black','AIAN','Asian','Hispanic','Other'])
```

This data is taken from:

- Census Racial Data

https://www.census.gov/quickfacts/fact/table/US/PST045219
Accessed Nov 3 2020


```python
# These represent number of people
total = 328239523
perc = np.array([0.601, 0.134, 0.013, 0.059, 0.185])
```

```python
# This determines the other percentage
1-np.sum(perc)
```

```python
perc = np.array([0.601, 0.134, 0.013, 0.059, 0.185, 0.008])
population = perc * total
population
# Population represents the number of people in different racial groups
```

```python
# Get COVID-19 deaths per 100,000
covid_rate = covid / population
per100 = covid_rate * 100000
per100
```

```python
# Create a bar graph that shows COVID-19 deaths
x_values = race
y_values = per100

fig, axs = plt.subplots(figsize=(10,6))      
axs.bar(x_values, y_values, color=('blue'), alpha=0.5)    
axs.set_title("COVID-19 Deaths Broken Down by Race", fontsize=20, fontweight="bold", pad=20)   
axs.set_xlabel("Race", fontsize=14)
axs.set_ylabel("COVID-19 Deaths per 100,000", fontsize=14)
axs.tick_params(labelsize=14)   

axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)

# Create a new variable that has the values rounded to 1 decimal place for textb
y_value1 = np.round(per100, decimals=1)

for i in range(len(x_values)):
    plt.text(i-0.2, y_value1[i]+1, str(y_value1[i]), color='b', size=16)

plt.grid(axis='y')
plt.text(-0.8, -20, "*Data Source: National Center for Health Statistics, U.S. Census Bureau", fontsize=12)

plt.show()
```

```python
# Get relative death rates
relative = per100 / per100[0]
relative
```

```python
fig, axs = plt.subplots(figsize=(10,6))
axs.barh(race, relative ,color=('#c1ba9d','#733d47','#733d47','#733d47','#733d47','#733d47'))
axs.invert_yaxis()
axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)
axs.spines['bottom'].set_visible(False)
axs.spines['left'].set_visible(False)
axs.get_xaxis().set_visible(False)
axs.set_yticklabels(race, size=15)
axs.set_title('Relative COVID-19 Death Rates', fontweight="bold", fontsize=20)

plt.text(0.85,0,'1.00',color='black',va="center",size=15, fontweight="bold")
plt.text(1.6,1,'1.74',color='white',va="center",size=15,fontweight="bold")
plt.text(0.75,2,'0.90',color='white',va="center",size=15,fontweight="bold")
plt.text(0.65,3,'0.79',color='white',va="center",size=15,fontweight="bold")
plt.text(1.15,4,'1.29',color='white',va="center",size=15,fontweight="bold")
plt.text(1.45,5,'1.61',color='white',va="center",size=15,fontweight="bold")

plt.text(0, 6,"*Data Source: National Center for Health Statistics, U.S. Census Bureau", fontsize=12 )

plt.show()
```
